# Informe del dataset de mujeres ajedrecistas

## Contexto
La Federación Internacional de Ajedrez (FIDE) gobierna la competencia internacional de ajedrez. La FIDE utilizó el sistema de clasificación Elo para calcular los niveles de habilidad relativos de los jugadores.

## Dataset
El dataset contiene detalles de las mejores jugadoras de ajedrez en el mundo ordenados por su calificación FIDE estándar (de mayor a menor por encima de 1800 Elo) según se actualizó en agosto de 2020. Los datos incluyen a todos los jugadores activos e inactivos que se pueden identificar mediante la columna Inactive_flag.

Nota: Todas las calificaciones se actualizan según lo publicado por FIDE en agosto de 2020.

## Referencias
FIDE: https://www.fide.com/
Kaggle: https://www.kaggle.com

### Importación de ibrerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Carga del dataset

```python
ruta = 'data/top_women_chess_players_aug_2020.csv'
df = pd.read_csv(ruta)
print(df)
```

### Información extra del dataset

```python
df.info()
```

### Explicacion de algunos valores
En la columna de titulos (**Title**) encontraremos abreviaciones de los títulos reconocido por la FIDE

    -GM: Grand Master
    -IM: International Master
    -FM: FIDE Master
    -CM: Candidate Master
Siendo el GM el título más alto al que se puede aspirar
 
Por otro lado las columnas de:
Standard_Rating  Rapid_rating  Blitz_rating

Representan el rating de las jugadores en diferentes modos de juego:

    -Estandar: 90 minutos para los primeros 40 movimientos seguidos de 30 minutos para el resto del juego
    -Rapid: Cada jugador dispone de 15 minutos más 10 segundos adicionales por cada jugada
    -Blitz: cada jugador tiene 3 minutos más 2 segundos adicionales por movimiento
    
Por último la columna de **Inactive_flag** tendra el valor de **wi** (woman inactive) en caso de que la jugadores no se encuentra activa caso contrario tendra NaN

## Preguntas

### 1. ¿Quién es la mejor ajedrecista según su rating FIDE?

```python
def barrasVertical(titulo, anchoBarra, n, y, etiquetaY, etiquetaX, limiteY,color):
    plt.figure(figsize=(14,8))
    plt.title(titulo)
    x = np.arange(n)
    plt.ylabel(etiquetaY)
    plt.xticks(x, (etiquetaX))
    plt.bar(x,y,anchoBarra, color= color)
    plt.ylim(limiteY)
    plt.show()
    
ancho = 0.35 
y = df[['Standard_Rating','Rapid_rating','Blitz_rating']]
y = y.loc[0]
etiquetaX = ['Standard_Rating','Rapid_Rating','Blitz_Rating']
limiteY=(2000,2800)
barrasVertical(df['Name'][0], ancho, 3, y, 'Rating Alcanzado', etiquetaX, limiteY, 'm')
```

#### Judit Polgar, es la mejor según su rating de las 3 modalidades de ajedrez, destacando principalmente en el modo Blitz

### 2. ¿Quienes lideran el top 10 de mejores ajedrecistas en modo Standard?

```python
def barrasHorizontal(titulo, dataFrame, etiquetaX, etiquetaY, limiteX):
    plt.figure(figsize=(14,8))
    plt.title(titulo)
    sns.barplot(x = etiquetaX, y = etiquetaY, data=dataFrame).set_xlim(limiteX)
    plt.show()

top_10 = df.head(10)
limiteX=(2400,2700)
barrasHorizontal("Top 10 mejores ajedrecistas Standard", top_10, 'Standard_Rating', 'Name', limiteX)
```

#### Como se puede ver las 10 mujeres que lideran el top en modo Standard apenas tienen diferencia en sus puntos, siendo la diferencia poco más de 200 puntos entre la décima jugadora y la primera

### 3. ¿Quienes lideran el top 10 de mejores ajedrecistas en modo Rapid?

```python
def barrasHorizontalesPlt(titulo, n, xValores, yValores, etiquetaY, limiteX, color):
    plt.figure(figsize=(14,8))
    plt.title(titulo)
    plt.ylabel(etiquetaY)
    y_pos = np.arange(n)
    plt.barh(y_pos, xValores, color = color)
    plt.yticks(y_pos, yValores)
    plt.xlim(limiteX)
    plt.show()

top_12 = df.head(12)
nombres = []
for nombre in (top_12['Name']):
    nombres.append(nombre)

ratings = []
for rating in (top_12['Rapid_rating']):
    ratings.append(rating)
limiteX = (2200,2700)
ratings.sort()
barrasHorizontalesPlt("Top 10 mejores ajedrecistas Rapid", len(nombres), ratings, nombres[2:], 'Nombres', limiteX, 'g')
```

#### Estas son las mejores jugadores en modo Rapid, Jun Xie y Susan Polgar no tienen datos oficiales en esta modalidad, y rellenar estos valores sería un error ya que solo la FIDE puede proporcionar los ratings de las ajedrecistas, por tanto se opto por incluir 2 ajedrecistas más para llenar el vacio de dichas jugadoras

### 4. ¿Quienes lideran el top 10 de mejores ajedrecistas en modo Blitz?

```python
top_12 = df.head(12)
nombres = []
top_12_final=top_12.dropna(subset = ['Blitz_rating'])
top_12_final.sort_values(by=['Blitz_rating'])
limiteX=(2400,2700)
barrasHorizontal("Top 10 mejores ajedrecistas Blitz", top_12_final, 'Standard_Rating', 'Name', limiteX)
```

#### Estas son las mejores jugadores en modo Blitz, las jugadoras que no tienen datos oficiales (FIDE), no fueron tomadas en cuenta para este top, como se puede observar en la gráfica la diferencia de puntos es muy poca habiendo menos de 200 puntos de diferencia entra la que lidera el top y la jugadora en la posicion 10

### 5. ¿Cual es la edad más común de las mejores ajedrecistas?

```python
def graficoDistribucion(data, ytitulo, xtitulo, titulo, stick):
    fig, ax = plt.subplots(figsize=(14,8))
    sns.distplot(data,  bins = 10, kde=False)
    ax.set_ylabel(ytitulo)
    ax.set_xlabel(xtitulo)
    if(stick):
        ax.set_xticks(range(0,101,10))
    plt.title(titulo)
    plt.show()

birth_year = df.Year_of_birth
birth_year=birth_year.dropna()
current_year = 2020
age_value =  current_year - birth_year

graficoDistribucion(age_value,"Jugadoras","Edad","Distribucion de edades de ajedrecistas",True)
```

#### Como se observa en el gráfico el rango de edad en el que se encuentran la mayoría de ajedrecistas profesionales es de 30 a 40 años, seguido por las ejedrecistas de 20 a 30 años

### 6. ¿En que año nacieron las mejores ajedrecistas?

```python
birth_year = df.Year_of_birth
birth_year=birth_year.dropna()
graficoDistribucion(birth_year,"Jugadoras","Año","Distribucion de fechas de nacimiento",False)
```

#### Como se observa en el gráfico el año en que nacieron la mayoria de ajedrecistas profesionales es en los años 80, seguido de los años 90

### 9. ¿Cuál es el título más común entre las profesionales?

```python
def graficoBarrasVertical(title, xValues, yValues, datos, xLabel, yLabel):
    fig, ax = plt.subplots(figsize=(14,8))
    plt.title(title)
    sns.barplot(x=xValues, y=yValues, data=datos)
    ax.set_xlabel(xLabel)
    ax.set_ylabel(yLabel)
    plt.show()

title_dist = df.Title.value_counts().reset_index()
graficoBarrasVertical("Distribución de titulos","index", "Title", title_dist,"Titulos","Jugadoras")
```

#### Como podemos ver el título más comun es WFM (Woman FIDE Master), seguido de WIM (Woman International Master)

### 10. ¿Cuál es el estatus actual de las mejores ajedrecistas?

```python
im = df["Inactive_flag"].value_counts(dropna=False)
df_status = pd.DataFrame({'labels': im.index,'values': im.values})
df_status=df_status.fillna('active')

graficoBarrasVertical("Estatus de ajedrecistas","labels", "values", df_status,"Estatus","Jugadoras")
```

#### Como muestra la gráfica el estatus de las ajedrecistas es mayormente wi (Woman Inactive), y las jugadores activas son aproximadamente la mitad de las inactivas

###  11. ¿Cuál es el estatus actual de las 10 mejores ajedrecistas?

```python
iwm = df.head(10)
iwm = iwm["Inactive_flag"].value_counts(dropna=False)
mejores_10 = pd.DataFrame({'labels': iwm.index,'values': iwm.values})
mejores_10=mejores_10.fillna('active')
graficoBarrasVertical("Estatus de ajedrecistas","labels", "values", mejores_10,"Estatus","Jugadoras")
```

#### Dentro del circulo de las 10 mejores ajedrecistas podemos ver que sucede lo contrario, ya que las jugadoras activas son 6 mientras que las inactivas 4, siendo mayoría las jugadores activas

### 12. ¿Cuál es el rating (standard) promedio por cada título?

```python
avg_std_rating_per_title = round(df.groupby("Title")["Standard_Rating"].mean(), 2).reset_index().sort_values(by='Standard_Rating', ascending=False).reset_index(drop=True)
xValues = "Title"
yValues="Standard_Rating"
graficoBarrasVertical("Rating Standard Promedio por título",xValues, yValues, avg_std_rating_per_title,"Titulos", "Promedio")
```

#### Como la gráfica muestra el Rating Standard promedio más alto se encuentra entre las GM (Grand Master), seguidas por las IM (International Master)
